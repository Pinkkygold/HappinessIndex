### Determining Normality


```python
df1 = df.copy()
```

```python
print("Missing value pattern:")
print(df1.isnull().sum())
```

```python
missing_rows = df1[df1.isnull().any(axis=1)]
print(f"Rows with missing values: {len(missing_rows)}")
print(missing_rows.head())
print("Percentage missing:", (len(df1[df1.isnull().any(axis=1)]) / len(df1)) * 100)
```

```python
df1 = df1.dropna() 
```

```python
df1.isnull().sum() 
```

```python
x = df1.drop(columns="MEDV")
y = df1['MEDV']
```

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression 

model = LinearRegression() 
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred 
```

```python
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) 

print(mse)
print(r2*100)
```
