```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore') 
```

```python
from sklearn.datasets import load_breast_cancer 

data = load_breast_cancer() 
df = pd.DataFrame(data.data, columns= data.feature_names)
df['Target'] = data.target 
df.head() 
```

```python
corr = df.corr(method='pearson') 
plt.figure(figsize=(10,8)) 
sns.heatmap(corr, annot=True, cmap='Greens') 
plt.show() 
```

```python
feature_name = 'mean radius'
x = df[[feature_name]].values
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 

model = LogisticRegression() 
model.fit(x_train, y_train) 
```

```python
y_pred = model.predict(x_test) 
y_pred 
```

```python
from sklearn.metrics import accuracy_score, classification_report 

print("Classification Report:\n", classification_report(y_pred, y_test))
print("Accuracy Report:", accuracy_score(y_pred, y_test)*100) 
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z)) 
```

```python
x_val = np.linspace(x.min(), x.max(), 569).reshape(-1,1) 
z_val = model.intercept_ + model.coef_[0][0] * x_val 
p_val = sigmoid(z_val) 
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x, y, alpha=0.25)
plt.plot(x_val, p_val, color='red')
plt.xlabel(feature_name)
plt.ylabel('Predicted probability')
plt.show()
```
